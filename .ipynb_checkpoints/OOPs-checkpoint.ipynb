```python
class Parking:
    
    def __init__(self, tot_tickets, tot_parking):
        self.tickets = list(range(1, tot_tickets + 1))
        self.parkingSpaces = list(range(1, tot_parking + 1))
        self.currentTicket = {}
        
    def takeTicket(self):
        if self.tickets:
            ticket = self.tickets.pop(0)
            space = self.parkingSpaces.pop(0)
            self.currentTicket[ticket] = {"paid" : False, "space" : space }
            print(f"Ticket number : {ticket}.")
            print(f"Parking lot : {space}.")
        else:
            print("Sorry, no more tickets!")
        
    def payForParking(self):
        ticket_number = int(input("Enter your ticket number"))
        if ticket_number in self.currentTicket:
            if self.currentTicket[ticket_number]["paid"] == False:
                payment = input("Enter the payment amount: ")
                if payment:
                    self.currentTicket[ticket_number]["paid"] = True
                    print("Payment received! You have 15 minutes to leave.")
                else:
                    print("Payment not received! Please pay to exit.")
            else:
                print("Ticket has already been paid.")
        else:
            print("Invalid ticket number.")
                    
    def leaveGarage(self):
        ticket_number = int(input("Enter your ticket number: "))
        if ticket_number in self.currentTicket:
            if self.currentTicket[ticket_number]["paid"]:
                self.parkingSpaces.append(self.currentTicket[ticket_number]["space"])
                self.tickets.append(ticket_number)
                print("Thank you, have a nice day!")
                del self.currentTicket[ticket_number]
            else:
                print("Payment not received. Please pay before leaving.")
        else:
            print("Invalid ticket number.")
    
```

```python
tot_tickets = 100  
tot_parking = 50  

parking_lot = Parking(tot_tickets, tot_parking)

while True:
    print("\n1. Take a ticket\n2. Pay for parking\n3. Leave the garage\n4. Quit")
    choice = int(input("Enter your choice (1-4): "))

    if choice == 1:
        parking_lot.takeTicket()
    elif choice == 2:
        parking_lot.payForParking()
    elif choice == 3:
        parking_lot.leaveGarage()
    elif choice == 4:
        print("Exiting the program. Goodbye!")
        break
    else:
        print("Invalid choice. Please enter a number from 1 to 4.")
```
